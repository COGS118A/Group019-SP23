```python
# We will use standard scaler from sklearn to transform numerical features, which include BMI, MentHlth, PhysHlth, and Age
from sklearn.preprocessing import StandardScaler

sdScaler = StandardScaler()
sdScaler.fit(df[['BMI', 'MentHlth', 'PhysHlth', 'Age']])
df[['BMI', 'MentHlth', 'PhysHlth', 'Age']] = sdScaler.transform(df[['BMI', 'MentHlth', 'PhysHlth', 'Age']])
df.head()
```

```python
# We will use onehot encoder from sklearn to transform categorical feature, age
from sklearn.preprocessing import OneHotEncoder

onehot = OneHotEncoder(handle_unknown='ignore')
onehot.fit(df[['Sex']])
df[onehot.get_feature_names_out(['Sex'])] = onehot.transform(df[['Sex']]).toarray()
df = df.drop(['Sex'], axis=1)
df.head()
```
